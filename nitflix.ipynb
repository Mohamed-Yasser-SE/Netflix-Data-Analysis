```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv(r"C:\Users\shiko\Desktop\projects\nitflix\netflix_titles.csv")
df.info()
```

```python
data=df[['type','director','title','rating','release_year','country']].dropna().reset_index()
data
```

```python
dup=data['title'].duplicated().sum()
print(f"sum of duplicates = {dup}")
```

```python
d=data['type'].value_counts()
plt.figure(figsize=(8,5))
plt.bar(d.index,d.values,color='black')
plt.grid(axis='y',color='b',alpha=0.5,linestyle=':')
plt.title('Number of movies and TV shows released from 1942 to 2021')
for i,v in enumerate(d):
    plt.text(i,v,v,ha='center',va='bottom')
plt.show()
```

```python
labels=(d.values/(5561+189)*100).round(1)
plt.pie(d,startangle=90,labels=labels,colors=['black','b'])
plt.legend(labels=d.index,loc='upper right')
plt.show()
```

```python
d2=data['release_year'].value_counts()
d2
```

```python
df2=pd.DataFrame({'year':d2.index,'values':d2.values})
plt.figure(figsize=(20,5))  
sns.barplot(x='year',y='values',data=df2,palette='flare')
plt.grid(axis='y',color='b',alpha=0.5,linestyle=':')
plt.xticks(rotation=90)
plt.title('Number of movies and TV shows each year')
plt.xlabel('YEAR')
plt.show()
```

```python
d3=data['director'].value_counts()
valuess=d3.loc[d3.values>6]
df3=pd.DataFrame({'Directors':valuess.index,'values':valuess.values})
plt.figure(figsize=(20,5))  
sns.barplot(x='Directors',y='values',data=df3,palette='crest')
plt.title('Directors who directed more than 6 movies and TV shows')
plt.xlabel('Directors')
plt.grid(axis='y',color='b',alpha=0.5,linestyle=':')
plt.xticks(rotation=90)
plt.show()
```
